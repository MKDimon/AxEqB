```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Данные
data = [
    [0.0147055, 0.0168759, 0.0081893, 0.0051943, 200],
    [0.0103852, 0.0103901, 0.0055912, 0.0029575, 200],
    [0.0101643, 0.0100876, 0.0048722, 0.0036332, 200],
    [0.0246027, 0.0294369, 0.0170098, 0.0100886, 300],
    [0.0246559, 0.0238129, 0.0154904, 0.0081722, 300],
    [0.023518, 0.0261907, 0.0144401, 0.0072813, 300],
    [0.0406721, 0.054276, 0.0255608, 0.0181842, 400],
    [0.0377798, 0.0607968, 0.025999, 0.0129962, 400],
    [0.0370596, 0.0465678, 0.0232791, 0.0113862, 400],
    [0.0583149, 0.0870657, 0.0363153, 0.0234688, 500],
    [0.0557055, 0.0951449, 0.0334027, 0.0208391, 500],
    [0.0517058, 0.0902606, 0.0318598, 0.014922, 500],
    [0.0793766, 0.102644, 0.04313, 0.02893, 600],
    [0.0727253, 0.106593, 0.0437337, 0.026938, 600],
    [0.0694905, 0.111364, 0.0410851, 0.0234512, 600],
]

# Переформатирование данных: усреднение каждой тройки строк, кроме последнего столбца
averaged_data = []
for i in range(0, len(data), 3):
    avg_row = np.mean(data[i:i+3], axis=0)
    averaged_data.append(avg_row)

# Создание DataFrame
columns = ['eigen', 'non-parallel', 'threads 2', 'threads 4', 'matrix_size']
df = pd.DataFrame(averaged_data, columns=columns)

# Группировка данных по размеру матрицы и вычисление среднего значения для каждой функции
grouped = df.groupby('matrix_size').mean()

# Построение графиков
plt.figure(figsize=(10, 6))
for col in ['eigen', 'non-parallel', 'threads 2', 'threads 4']:
    plt.plot(grouped.index, grouped[col], marker='o', label=col)

plt.title('Зависимость времени вычисления от размера матрицы')
plt.xlabel('Размер матрицы')
plt.ylabel('Время вычислений')
plt.legend()
plt.grid()
plt.show()

```
